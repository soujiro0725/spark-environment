```python
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.models import Sequential
import lstm, time #helper libraries
```

```python
X_train, y_train, X_test, y_test = lstm.load_data('sp500.csv', 50, True)
```

```python
model = Sequential()
model.add(LSTM(
    input_dim=1,
    output_dim=50,
    return_sequences=True
))
model.add(Dropout(0.2))
model.add(LSTM(
    100,
    return_sequences=False
))
model.add(Dropout(0.1))
model.add(Dense(
    output_dim=1
))
model.add(Activation('linear'))
start = time.time()
model.compile(loss='mse', optimizer='rmsprop')
print('compilation time :', time.time() - start)
```

```python
model.fit(
X_train,
y_train,
batch_size=512,
nb_epoch=1,
validation_split=0.05)
```

```python
predictions = lstm.predict_sequences_multiple(model, X_test, 50, 50)
lstm.plot_results_multiple(predictions, y_test, 50)
```

```python
import numpy as np
```

```python
data = np.random.rand(250, 5)
labels = (np.sum(data, axis=1) > 2.5) * 1
```

```python
labels
```

```python
from keras.utils import np_utils
```

```python
labels =  np_utils.to_categorical(labels)
```

```python
labels.shape
```

```python
model = Sequential()
```

```python
model.add(Dense(20, input_dim=5))
```

```python
model.add(Activation('relu'))
```

```python
model.add(Dense(2, activation='softmax'))
```

```python
model.compile('rmsprop', 'categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(data, labels, nb_epoch=300, validation_split=0.2)
```

```python
test = np.random.rand(200, 5)
predict = np.argmax(model.predict(test), axis=1)
real = (np.sum(test, axis=1) > 2.5) * 1
sum(predict == real) / 200.0
```
